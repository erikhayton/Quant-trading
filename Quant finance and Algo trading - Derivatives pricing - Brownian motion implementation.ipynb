```python
import math
import numpy as np
import numpy.random as npr
import scipy
import matplotlib.pyplot as plt
```

```python
def brownian_motion(dt = 0.1, X0=0, N=1000):
    #initialise W(t) with zeros
    W = scipy.zeros(N+1)
    
    #creating N+1 time steps
    t = scipy.linspace(0,N, N+1)
    
    #cumulatuve sum on every step draws from a normal distribution with mean 0 and variane dt N(0,dt)
    W[1:N+1] = scipy.cumsum(scipy.random.normal(0,dt,N))
    
    return t, W  
    
```

```python
def plot_brownian_motion(t,W):
    
    plt.plot(t,W)
    plt.xlabel('Time (t)')
    plt.ylabel('Weiner-processL W(t)')
    plt.title('Wiener-process')
    plt.show()
```

```python
if __name__ == '__main__':
    t,W = brownian_motion()
    plot_brownian_motion(t,W)
```
