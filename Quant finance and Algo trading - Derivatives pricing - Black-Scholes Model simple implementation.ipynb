```python
def blackscholes_call(S, E, T, rf, sigma):
    #Calculate d1 and d2
    d1 = (log(S/E)+(rf+ sigma*sigma/2.0)*T)/(sigma*sqrt(T))
    d2 = d1 - sigma*sqrt(T)
    print(round(d1,3))
    print(round(d2,3))
    
    return S* stats.norm.cdf(d1)- E*exp(-rf*T)*stats.norm.cdf(d2)
```

```python
def blackscholes_put(S,E,T,rf,sigma):
    #calculate d1 and d2
    d1 = (log(S/E) + (rf+sigma*sigma/2.0)*T)/ (sigma*sqrt(T))
    d2 = d1 - sigma*sqrt(T)
    
    return -S*stats.norm.cdf(-d1) + E*exp(-rf*T)*stats.norm.cdf(-d2)
```
